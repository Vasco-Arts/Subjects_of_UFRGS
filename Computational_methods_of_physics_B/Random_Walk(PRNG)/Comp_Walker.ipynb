**$\underline{Comparison\hspace{0.2cm} between\hspace{0.2cm} Python\hspace{0.2cm} Random\hspace{0.2cm} and\hspace{0.2cm} Math\hspace{0.2cm} Method}$**

```python
import numpy as np
import matplotlib.pyplot as plt
import random

a,b,c = 1029,221591,1048576
seed = 3072
N_W = 1 #1000 # Number of walker (or hikers). 
N_S = 10000 #10000 # Number of steps.

''' Linear Congruential Generator:'''
def rand():
  global seed
  seed = (a * seed + b)%c
  return float(seed/c)

def coin():
  x = 2 * int(2 * rand()) - 1
  return x

for j in range(N_W):
  walk = 0
  list_t = []; list_walk = []
  list_t.append(0);list_walk.append(walk)
  
  for t in range(1,N_S):
    walk += coin() # We are adding walk = walk + coin()
    list_t.append(t)
    list_walk.append(walk) 
  plt.plot(list_t,list_walk) # Plot each graph.
#------------------------------------------------------------------------

''' Python Generator:'''
def coin2():
  x = 2 * int(2 * random.uniform(0,1)) - 1
  return x

for j in range(N_W):
  walk = 0
  list_t2 = []; list_walk2 = []
  list_t2.append(0);list_walk2.append(walk)
  
  for t in range(1,N_S):
    walk += coin2()
    list_t2.append(t)
    list_walk2.append(walk)
  plt.plot(list_t2,list_walk2)

plt.title('Graph of Random Walk')
plt.legend(['Linear Congruential Generator method','Python Library'])
plt.xlabel('Steps')
plt.ylabel('Down/Up or Left/Right')
plt.show()
```
