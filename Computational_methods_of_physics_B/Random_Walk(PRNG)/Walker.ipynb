**$\underline{Random\hspace{0.2cm} Walk}$**

We simulate the path that a person can take with random steps to one side or the other.

Similar to the coin algorithm, the walker has a 50% probability of walking up or down (left or right, head or tail, 1 or -1, 1 or 0 bit, etc), and its previous step does not interfere with the next step.

With two walkers:

```python
import numpy as np
import matplotlib.pyplot as plt

a,b,c = 1029,221591,1048576
seed = 3072
t,j = 0,0
N_W = 2 #1000 # Number of walker (or hikers). 
N_S = 1000 #10000 # Number of steps.

def rand():
  global seed
  seed = (a * seed + b)%c
  return float(seed/c)

def coin():
  x = 2 * int(2 * rand()) - 1
  return x

# Loop of Random Walk:
for j in range(N_W):
  walk = 0
  list_t = []; list_walk = []
  list_t.append(0);list_walk.append(walk)
  
  for t in range(1,N_S):
    walk += coin() # We are adding walk = walk + coin()
    list_t.append(t)
    list_walk.append(walk)
  
  plt.plot(list_t,list_walk) # Plot each graph.
plt.title('Graph of Random Walk')
plt.xlabel('Steps')
plt.ylabel('Down/Up or Left/Right')
plt.show() # Show all graphs together.
```
