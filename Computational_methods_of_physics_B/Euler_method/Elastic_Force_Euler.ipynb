**$\underline{Elastic\hspace{0.2cm} Force\hspace{0.2cm} with\hspace{0.2cm} Explicit\hspace{0.2cm} Euler}$**

This example is for us to proof that Euler's method is not always precise (or correct), in this case, we can see the curves increase their amplitude. But this cannot occur, because of the energy conservation (here we got a close system, with only conservative force), the amplitude must be constant (its a simple harmonic oscillator). 

The explanation of this is, the fact of time interval is a linear variation makes inaccurate in a curve variation of time (because is linear), as in an oscillation. This can be better observed with a large frequency (small wavelength), because at every curve, the error will be increased. So if we put a lot of curves, we will see the amount of the errors being summed. 

The smaller $\Delta t$, smaller the error, because we can "cut" a curve in a lot of straight lines. 

We can compare with methods of Leap-Frog, Verlet and Velocity-Verlet (in other codes)(there is a code that compares these three methods), where there is no error in the same example.

```python
import numpy as np
import matplotlib.pyplot as plt
# The situation is: an object pulled by a spring, the initial value is v = 0 (of the block) and x = 1. 

v = 0 # Initial velocity v0.
x = 1 # Initial position x0, in centimeters.
dt = 0.0005 # Time interval, step in which time passes (interval between t1 and t2).
m = 1 # Spring mass.
k = m # Spring constant.   
w = k/m # = 1 Angular frequency.       
y = x # Auxiliary variable.  
t = 0 # Initial time.
tf = 500 # Final time, in minutes.

def a(x): # Spring acceleration function.
  a = -w**2 * x # a(x) = (-kx)/m = -wx Acceleration of a elastic spring.
  return a

# Lists:
list_t = [t]
list_x = [x]
list_v = [v]

while t < tf:
  x = x + v * dt
  list_x.append(x)
  #list_x = np.array(list_x)
  
  #v = v - w**2 * y * dt
  v = v + a(y) * dt
  list_v.append(v)
  #list_v = np.array(list_v)

  y = x # Auxiliary variable. 
  t = t + dt # Time count.
  list_t.append(t)
  #list_t = np.array(list_t)


plt.plot(list_t,list_v)
plt.plot(list_t,list_x)
plt.title('Oscillation of an Elastic Force, with Euler')
plt.xlabel('Time (minutes)')
plt.ylabel('Position (cm)')
plt.show()
```
