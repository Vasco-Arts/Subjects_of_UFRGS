**$\underline{Comparison\hspace{0.2cm} between\hspace{0.2cm} Verlet,\hspace{0.2cm} Leap-Frog\hspace{0.2cm} and\hspace{0.2cm} Velocity-Verlet}$**

Here, we will compare these three methods with the example of elastic force, again.

We expect the results to be the same, so the lines must be overlap. 

Note that the Velocity-Verlet is the only one that doesn't have a step outside the loop.

```python
import numpy as np
import matplotlib.pyplot as plt
'''      Showing the similarity between these methods: VELOCITY-VERLET, VERLET, LEAP-FROG            '''

# The situation is: an object pulled by a spring, the initial value is v = 0 (of the block) and x = 1.

# v1, x1 and x2 are from the Verlet method. 
# V3 and x3 are from the Leap Frog (LF) method. 
# V4 and x4 are from the Velocity Verlet method (vel_ver).

v1,v3,v4 = 0,0,0 
x1,x2,x3,x4 = 1,1,1,1
dt = 0.0005 
m = 1
k = m 
w = k/m
t = 0 
tf = 500
v1aux = 0 # verlet
v3med = 0 # LF 
v3aux = 0 # LF

# Lists:
list_t = [t]
list_x1 = [x2] # verlet
list_v1 = [v1] #verlet
list_x3 = [x3] # LF
list_v3med = [v3med] # LF
list_x4 = [x4] #vel_ver
list_v4 = [v4] #vel_ver

def euler(x0,v,dt): # Verlet
  x = x0 + v * dt 
  return x

x2 = euler(x1,v1,dt) # Verlet

def a(x):
  a = -(w**2) * x 
  return a 

v3 = v3 + a(x3) * (dt/2) # LF

while t < tf: 
  # Verlet method:

  x1aux = x2 # verlet
  x2 = 2 * x2 - x1 + a(x2) * dt**2 # verlet
  list_x1.append(x2)

  v1 = (x2 - x1)/(2 * dt) # verlet
  list_v1.append(v1)

  x1 = x1aux # verlet
  

  # Leap-frog method:
  
  x3 = x3 + v3 * dt # x(dt) = x0 + v(dt/2) * dt
  list_x3.append(x3)
  
  v3med = v3 # LF
  
  v3 = v3 + a(x3) * dt # lF
  
  v3aux = v3 # New v, in this case v(dt + dt/2).
  
  v3med = ((v3med + v3aux)/2) # velocity average.
  list_v3med.append(v3med)


  # Velocity-Verlet method:

  v4 = v4 + a(x4) * dt/2

  x4 = x4 + v4 * dt
  list_x4.append(x4)

  v4 = v4 + a(x4) * dt/2
  list_v4.append(v4)

  t = t + dt 
  list_t.append(t)

plt.plot(list_t,list_v1)
plt.plot(list_t,list_x1)
plt.plot(list_t,list_v3med)
plt.plot(list_t,list_x3)
plt.plot(list_t,list_v4)
plt.plot(list_t,list_x4)
plt.title('Oscillation of an Elastic Force')
plt.legend(['v verlet','x verlet','v LF','x LF','v Vel-Ver','x Vel-Ver']) 
plt.ylabel('Position (cm)')
plt.xlabel('Time t (minutes)')
plt.show()

# It's not possible to see the other lines because the line of one method is above others.
```
