**$\underline{Elastic\hspace{0.2cm} Force\hspace{0.2cm} with\hspace{0.2cm} Leap\hspace{0.2cm} Frog}$**

In this code, we can see that the error using Euler's method decrease with the Leap Frog method. Remembering, because of the energy conservation, the amplitude of the oscillation must be constant, unlike the Euler's method.

In this method, we take the midpoint between $x(t + \Delta t)$ and $x(t)$

General equation: $f(t+\Delta t) = f(t) + f'(t + \Delta t/2)\Delta t$


-In the loop:

$v(t + \Delta t/2) = v(t - \Delta t/2) + a(t)\Delta t$

$x(t+\Delta t) = x(t) + v(t + \Delta t/2)\Delta t$

- Algorithm:


```
v = v + a(t) * dt/2 

for i in range()

  x = x + v*dt # x(\Delta t) = x(0)+v(\Delta t/2)*dt; 0 -> \Delta t
  
  v = v+a(t)*dt  # v(\Delta t + \Delta t/2) = v(\Delta t/2) + a(\Delta t)*dt; \Delta t/2 -> \Delta t+\Delta t/2




```
We want in the graph $v(\Delta t)$, to do this, we need to take the average: 

```

for i in range()

  v = v+a(t)*dt  # v(\Delta t + \Delta t/2) = v(\Delta t/2) + a(\Delta t)*dt; \Delta t/2 -> \Delta t+\Delta t/2

```







````python
import numpy as np
import matplotlib.pyplot as plt

v = 0 # This is the initial velocity of the system (v0).
x = 1 # Initial x0.
dt = 0.0005 # Step in which the time passes (interval between t1 and t2).
m = 1 # String mass.
k = m # String constant.
w = k/m # = 1 Angular frequency.
t = 0 # Time count.
tf = 500
v_ave = 0 # Average velocity.
v_aux = 0 # Auxiliary variable.

# Lists:
list_t = [t]
list_x = [x]
list_v_ave = [v_ave]

def a(x):
  a = -(w**2) * x # a(x) = (-k*x2)/m = -(w**2 * x2); k/m = w^2.
  return a 

v = v + a(x) * (dt/2) # This is v(dt/2) = v0.

while t < tf: # Total time that the event will occur in the system (time of the loop).
  x = x + v * dt # x (dt) = x0 + v (dt/2)*dt.
  list_x.append(x)

  v_ave = v # v(dt/2).
  
  v = v + a(x) * dt # v(dt+dt/2) = v(dt/2) + a(x(dt)) * dt.

  v_aux = v # New v, in this case v(dt+dt/2).
  
  v_ave = ((v_ave + v_aux)/2) # The average of the velocities.
  list_v_ave.append(v_ave)

  t = t + dt # Time passes equation, t passes with a length dt.
  list_t.append(t)

plt.plot(list_t,list_v_ave)
plt.plot(list_t,list_x)
#plt.legend(['velocity','distance']) 
plt.title('Oscillation of an elastic force')
plt.ylabel('Position (cm)')
plt.xlabel('Time t (minutes)')
plt.show()

````
