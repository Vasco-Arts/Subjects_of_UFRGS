````python
import numpy as np
import matplotlib.pyplot as plt
# The situation is: an object pulled by a spring, the initial value is v = 0 (of the block) and x = 1. 

v = 0 # Initial velocity v0.
x = 1 # Initial position x0, in centimeters.
dt = 0.0005 # Time interval, step in which time passes (interval between t1 and t2).
m = 1 # Spring mass.
k = m # Spring constant.   
w = k/m # = 1 Angular frequency.         
t = 0 # Initial time.
tf = 500 # Final time, in minutes.

# Lists:
list_t = [t]
list_x = [x]
list_v = [v]

def a(x): # Spring acceleration function.
  a = -(w**2) * x # a(x) = (-kx)/m = -wÂ²x Acceleration of a elastic spring.
  return a 


while t < tf: 
  v = v + a(x) * dt/2 # v(t + dt/2) = v(dt) + a(x(dt)) * dt/2

  x = x + v * dt # x (t + dt) = x(t) + v (t + dt/2) * dt
  list_x.append(x)

  # a(x) * (dt/2) # a(t + dt) * dt/2
  
  v = v + a(x) * dt/2 # v(t + dt) = v(t + dt/2) + a(x(dt)) * dt/2
  list_v.append(v)

  t = t + dt 
  list_t.append(t)

plt.plot(list_t,list_v)
plt.plot(list_t,list_x)
#plt.legend(['Velocity','Position'])
plt.title('Oscillation of an Elastic Force, with Velocity-Verlet')
plt.ylabel('Position (cm)')
plt.xlabel('Time t (minutes)')
plt.show()
````
