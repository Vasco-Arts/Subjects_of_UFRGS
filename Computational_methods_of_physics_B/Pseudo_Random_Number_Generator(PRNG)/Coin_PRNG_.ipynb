**$\underline{Random\hspace{0.2cm} Number\hspace{0.2cm} Generator}$**

This method generates pseudo random number, that is, numbers that appear to be random, but are not. A proof of this is that we can calculate by ourselves (using a mathematical algorithm), predicting the random values, and if we can do that, then it's not random. If we run this Pseudo Random Number Generator on another computer, the result will be the same, showing once again that this is not a random generator. This is because the initial conditions are always the same, the numbers involved in the algorithm do not change, so the result will always be the same. We can try to create complexities in the algorithm or we can try to change the parameters to simulate a randomness,but at some point the values ​​will start to repeat, to change that, this modification has to happen all the time, the moment it stops, it will become like in the beginning, but more complex. 

In C language at least, the random function uses the computer's clock time as a seed to generate random numbers, because time doesn't stop, however after 24 hours time repeats itself, and the random sequence repeats itself and becomes predictable. 

There is even a discussion in mathematics if there are random numbers or we just don't know the end of the predictable sequence (in mathematical methods).

$seed = (a.seed + b) \% c \Rightarrow$ seed $=$ Is the initial seed that will generate the random branches.

$\%=$ Is the remainder of the division.

$c$ = To improve randomness, the $c$ represents the number of possible values ​​(ie (that is), on a dice there are 6 sides, 6 values, so $c = 6$). Since it takes the remainder of the division, it is better to choose odd parameters of a,b,c, in addition to choosing very high numbers.

Our method generates numbers between 0 and 1, to generate numbers in larger ranges, it's just multiply, for example, by 1000, to generate values ​​between 0 and 1000.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

a,b,c = 1029,221591,1048576 # Fixed numbers.
seed = 3072
t,tf = 0,10

def rand(): # Function that generates pseudo random numbers between 0 and 1.
  global seed # Will generate a global/general variable that can be used in the rest of the code, not just inside this 
  # function, it's more or less like the #define of the C language.
  seed = (a * seed + b)%c
  return float(seed/c) # To force the result to be float, and prevent python from rounding to int.

print('Random numbers between 0 and 1:\n')
while t < tf:
  print('%d)'%(t + 1),rand())
  t = t + 1

# Generator between 0 and 1000:
print('\nRandom numbers between 0 and 1000:\n')
t,tf = 0,10
while t < tf:
  print('%d)'%(t + 1), rand() * 1000)
  t = t + 1
```

**Graph of Random Number:**

```python
import numpy as np
import matplotlib.pyplot as plt
#                                             Graph
a,b,c = 1029,221591,1048576 
seed = 3072 
t,tf = 0,300
list_Y,list_T = [],[]

def rand():
  global seed
  seed = (a * seed + b)%c
  return seed

while t < tf:
  y = float(rand()/c)
  list_Y.append(y)

  t = t + 1
  list_T.append(t)

plt.plot(list_T,list_Y)
plt.show()
plt.scatter(list_T,list_Y)
plt.show()
```

# Using this method of random number to toss some coins (randomly generate two numbers):

```python
import numpy as np
import matplotlib.pyplot as plt

# Here, the code will show the occurrence of heads and tails.
a,b,c = 1029,221591,1048576
x1 = 3072
t,tf = 0,50
list_x = []
list_t = []

def rand():
  global x1
  x1 = (a * x1 + b)%c
  return float(x1/c)

def coin(): # Function of the coin.
  x = 2 * int(2 * rand()) - 1 # Returns only 1 (heads) or -1 (tails), we can think in 1 and 0 for bits, etc.
  return x

while t < tf:
  x = coin()
  list_x.append(x)

  t = t + 1
  list_t.append(t)

plt.plot(list_t,list_x)
plt.show()

plt.scatter(list_t,list_x)
plt.show()
```
